```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("/Users/mustajabmalik/Downloads/cricket_data.csv")
```

```python
df.head(5)
```

```python
runs_scored = df.groupby("Player")["Runs"].sum()
print(runs_scored)
```

```python
runs_scored.plot(kind="bar", color = "teal")
plt.title("How many runs scored")
plt.show()
```
