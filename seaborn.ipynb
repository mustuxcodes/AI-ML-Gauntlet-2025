```python
#Seaborn is a library built on top of Matplotlib. It is designed to create beautiful and informative statistical graphics with very little code. It is the preferred tool for rapid exploratory data analysis.
                                                                                                                                                                        
```

```python
import seaborn as sns
```

```python
#basic plotting with seaborn
tips = sns.load_dataset("tips")
tips.head(5)
```

```python
#create a scatter plot using the above dataset
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y="tip", data=tips)
plt.title("total_bill vs tips")
```

```python
#lineplot using seaborn

sns.lineplot(x= "size", y="total_bill", data = tips)
plt.title("Total bill by size")
plt.show()
```

```python
## categorical plots
#barplot

sns.barplot(x= "day", y="total_bill", data=tips)
```

```python
## categorical plots
#barplot

sns.barplot(x= "time", y="total_bill", data=tips)
```

```python
tips.head(5)
```

```python
## categorical plots
#barplot

sns.barplot(x= "sex", y="total_bill", data=tips)
```

```python
#Boxplot --> it is easier to identiy outliers in this just like see below the graph you can easily identify outliers

sns.boxplot(x= "day", y="total_bill", data=tips)
plt.show()
```

```python
##viloin plot

sns.violinplot(x= "day", y="total_bill", data=tips)
plt.show()
```

```python
#Histograms

sns.histplot(tips ["total_bill"], bins=4, kde=True)
```

```python
#Histograms

sns.histplot(tips ["total_bill"], bins=4, kde=False)
```

```python
#kernal distrubution plot(kde)

sns.kdeplot(tips["total_bill"], shade= True)
```

```python
#kernal distrubution plot(kde)

sns.kdeplot(tips["total_bill"], shade= False)
```

```python
#kernal distrubution plot(kde)

sns.kdeplot(tips["total_bill"], fill= False)
```

```python
#kernal distrubution plot(kde)

sns.kdeplot(tips["total_bill"], fill= True)
```

```python
#parrplot --> relationship of each and every variable with each other

sns.pairplot(tips)
```

```python
#heatmap

corr = tips[["total_bill", "size", "tip"]].corr()
corr
```

```python
sns.heatmap(corr)
```
