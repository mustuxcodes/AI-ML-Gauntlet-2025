```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y= [2,6,8,10]

#create a lineplot
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.title("Practing my ml engineer journey")
plt.show()
```

```python
#customsing a plot
x = [1,2,3,4]
y= [2,6,8,10]

plt.plot(x,y,color = "green",linestyle = "--", markersize=7, linewidth = 4)
plt.grid(True)
```

```python
#plotting multiple plots hhere

x = [1,2,3,4]
y = [2,6,8,10]
z= [10,10,30,40]

plt.figure(figsize = (9,5))

plt.subplot(1,2,1)
plt.plot(x,y,color = "green")
plt.title("plot1")
plt.show()

plt.subplot(2,2, 2)
plt.plot(y,z, color="blue")
plt.title("This is second one")
plt.show()

plt.subplot(2,2, 3)
plt.plot(z,y, color="blue")
plt.title("This is 3rd one")
plt.show()
```

```python
#bar plot now

categories = ['A', 'B', 'C', 'D']
values = [10,20,30,40]

plt.bar(categories, values, color="Purple")
plt.xlabel("Values")
plt.ylabel("Idk random")
```

```python
#histograms= represents the distribution of dataset. They divide the data into bins and then and counts the number of data points in each bin

data=[1,2,4,5,6,78,88,8,8,92,12,12,12,34,35,6658,878]
#create a histogram

plt.hist(data,bins=4,color ="Orange",edgecolor ="Blue")
#plt.show()

```

```python
#scatter plot now
x = [1,2,3,4]
y = [2,6,8,10]

plt.scatter(x,y)
```

```python
#pie chart

labels = ['A','B','C','D']
sizes = [10,20,30,40]
colors= ["Pink", "lightgreen", "Lightblue", "Green"]
explode = (0.2,0,0,0) #move out the first slice code is in the next 

plt.pie(sizes, labels = labels, colors=colors)
plt.show()

```

```python
#pie chart

labels = ['A','B','C','D']
sizes = [10,20,30,40]
colors= ["Pink", "lightgreen", "Lightblue", "Green"]
explode = (0.2,0,0,0) #move out the first slice 

plt.pie(sizes,explode=explode, labels = labels, colors=colors, autopct ="%1.1f%%", shadow = True)
plt.show()

```
